```python
import pandas as pd, os, sys, ast, glob

# Import library with current code functions
sys.path.append(os.path.join("..", "lib"))
import manual_labeler_functions as man_lab_fun
```

```python
SOURCE_FOLDER = os.path.join('..', 'Dataset')

VD_LABEL_FILE_NAME = 'VD_LABELED_L0.CSV'

SOURCE_FOLDER_LOCAL = os.path.join(SOURCE_FOLDER, 'DD-Local')
FOLDER_PATH = os.path.join(SOURCE_FOLDER_LOCAL, '**', VD_LABEL_FILE_NAME)
MAIN_LIST_local = sorted(glob.iglob(FOLDER_PATH, recursive=True ))

SOURCE_FOLDER_YT = os.path.join(SOURCE_FOLDER, 'YT-Online')
FOLDER_PATH = os.path.join(SOURCE_FOLDER_YT, '**', VD_LABEL_FILE_NAME)
MAIN_LIST_YT = sorted(glob.iglob(FOLDER_PATH, recursive=True ))

# Join the Two Lists
MAIN_LIST_LABELER = MAIN_LIST_local + MAIN_LIST_YT

SOURCE_FOLDER_SEED = os.path.join('..', 'Dataset', 'REF-Gold-Label')
FOLDER_PATH = os.path.join(SOURCE_FOLDER_SEED, '**', VD_LABEL_FILE_NAME)
REFERENCE_LST = sorted(glob.iglob(FOLDER_PATH, recursive=True ))
```

```python
REFERENCE_LST
```

```python
path_file = REFERENCE_LST[0]
path_dir = os.path.dirname(path_file)
vd_labeled_path = os.path.join(path_dir, VD_LABEL_FILE_NAME)
vd_labeled = pd.read_csv(vd_labeled_path)
vd_labeled.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
vd_labeled
```

## Gets all saved label classes.

```python
all_class = man_lab_fun.GET_ALL_CLASSES (vd_labeled)
label_name = all_class[0]
reference_measures = man_lab_fun.GET_MEASURES_FROM_CLASS(vd_labeled, label_name)
frames = man_lab_fun.GET_FRAMES_FROM_CLASS (vd_labeled, label_name)
```

```python
VECTOR_SUBSERIES_DT = pd.DataFrame()
for current_path_location in MAIN_LIST_LABELER:
    VD_LABELED_DT = pd.read_csv(current_path_location)
    VD_LABELED_DT.drop(columns=['Unnamed: 0'], inplace=True)
    
    label_info_column = list(VD_LABELED_DT['label_inf'])
    init_end_set = set(label_info_column)
    for item in init_end_set:
        label_info_dict = ast.literal_eval(item)
        if len(label_info_dict) != 0:
            init, end = label_info_dict[label_name]
            VD_LABELED_DT = VD_LABELED_DT.set_index(pd.Index(VD_LABELED_DT['frame_seq']))
            TEMP_DT = VD_LABELED_DT.loc[init:end].reset_index(drop=True)
            VECTOR_SUBSERIES_DT = pd.concat([VECTOR_SUBSERIES_DT, TEMP_DT])

print('Saving file...')
OUTPUT_NAME = 'VECTOR_SUBSERIES_DT.CSV'
VECTOR_SUBSERIES_DT.to_csv(OUTPUT_NAME)
```

```python
# Select the labeled subseries
Pattern_DT = vd_labeled.loc[frames]
Pattern_DT = Pattern_DT[reference_measures]
```

```python
(Pattern_DT[reference_measures].values)
```

## Prepare for the neural network

```python
for measure in reference_measures:
    VECTOR_SUBSERIES_DT = pd.DataFrame()
    for i, current_path_location in enumerate(MAIN_LIST_LABELER):
        VD_LABELED_DT = pd.read_csv(current_path_location)
        VD_LABELED_DT.drop(columns=['Unnamed: 0'], inplace=True)
        VD_LABELED_DT = VD_LABELED_DT.set_index(pd.Index(VD_LABELED_DT['frame_seq']))

        label_info_column = list(VD_LABELED_DT['label_inf'])
        init_end_set = set(label_info_column)
        for item in init_end_set:
            label_info_dict = ast.literal_eval(item)
            if len(label_info_dict)!=0:
                init, end = label_info_dict[label_name]
                TEMP_DT = VD_LABELED_DT.loc[init:end].reset_index(drop=True)
                TEMP_DT = TEMP_DT[measure]
                TEMP_DT = TEMP_DT.rename(i+1)
                VECTOR_SUBSERIES_DT = pd.concat([VECTOR_SUBSERIES_DT, TEMP_DT], axis=1)

    print('VECTOR_SUBSERIES_DT', VECTOR_SUBSERIES_DT)
    print('Pattern_DT[measure].values', Pattern_DT[measure].values)

    # If 'Reference' is not in VECTOR_SUBSERIES_DT.columns: 
    VECTOR_SUBSERIES_DT.insert(0, 'Reference', Pattern_DT[measure].values)

    OUTPUT_NAME = f"lista_padroes_interesse_{measure}.csv"
    VECTOR_SUBSERIES_DT.to_csv(OUTPUT_NAME)        
```
