# This notebook show all indexed videos information

```python
import os
import sys
import pandas as pd

# Import library with current code functions
sys.path.append(os.path.join("..", "lib"))
import general_functions as gf, files_paths as fp
```

## Search all Dataset VD_INFO files

```python
VD_INFO_LIST = gf.find_files_in_all_subdirectories([fp.DATASET_LOCAL, fp.DATASET_YT], fp.VD_INFO)
VIDEO_INDEX_DTFRAME = pd.DataFrame()
```

## Concatenate all the VD_INFOS information

```python
for current_path_location in VD_INFO_LIST:
    video_info = gf.READ_CSV_FILE(current_path_location)

    video_feature = gf.READ_CSV_FILE(os.path.dirname(current_path_location), fp.VD_FEATURES_L1)
    
    # Count instances in each video
    video_info['total_instance'] = video_feature['instance_ref'].nunique()
    
    # Mount index dataframe
    VIDEO_INDEX_DTFRAME = pd.concat([VIDEO_INDEX_DTFRAME, video_info], axis=0, ignore_index=True)
```

## Show the number of each origin video

```python
print('Total DD-Local Videos:', len(VIDEO_INDEX_DTFRAME.query("origin_vid == 'D'")))
print('Total YT-Online Videos:', len(VIDEO_INDEX_DTFRAME.query("origin_vid == 'Y'")))
```
